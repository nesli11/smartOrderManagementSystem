```python
#HAM SİPARİŞ VERİSİ
```

```python
orders = [
    {
"customer":"Ahmet",
"items": ["Laptop","Mouse","Mouse"],
"total":18500,
"is_paid":True
    },
    {
"customer":"Zeynep",
"items": ["Keyboard"],
"total":2500,
"is_paid":False
    },
    {
"customer":"Mehmet",
"items": ["Monitor","HDMI Cable","Monitor"],
"total":7200,
"is_paid":True
    },
    {
"customer":"Mehmet",
"items": ["Monitor","HDMI Cable","Monitor"],
"total":7200,
"is_paid":True
    }
]

```

```python
#SİPARİŞ SAYISI ANALİZİ(döngü kullanmadan)
```

```python
toplam_siparis = len(orders)
```

```python
odenmis_siparis = 0
```

```python
odenmemis_siparis = 0
```

```python
if orders[0]["is_paid"] == True :
    odenmis_siparis += 1
else :
    odenmemis_siparis += 1
```

```python
if orders[1]["is_paid"] == True :
    odenmis_siparis += 1
else :
    odenmemis_siparis += 1
```

```python
if orders[2]["is_paid"] == True :
    odenmis_siparis += 1
else :
    odenmemis_siparis += 1
```

```python
if orders[3]["is_paid"] == True :
    odenmis_siparis += 1
else :
    odenmemis_siparis += 1
```

```python
print(f"toplam sipariş sayısı : {toplam_siparis}")
print(f"ödenmiş sipariş sayısı : {odenmis_siparis}")
print(f"ödenmemiş sipariş sayısı : {odenmemis_siparis}")
```

```python
#MÜŞTERİ LİSTESİ
```

```python
musteri_listesi = [orders[0]["customer"],orders[1]["customer"],orders[2]["customer"],orders[3]["customer"]]
```

```python
temizlenmis_musteri_listesi = set(musteri_listesi)
```

```python
print(f"müşteri listesi : {musteri_listesi}")
print(f"temizlenmiş müşteri listesi : {temizlenmis_musteri_listesi}")
```

```python
#ÜRÜN ANALİZİ
```

```python
tum_urunler = (orders[0]["items"] + orders[1]["items"] + orders[2]["items"] + orders[3]["items"])
```

```python
temizlenmis_urunler = set(tum_urunler)
```

```python
print(f"tüm siparişler : {tum_urunler}")
print(f"temizlenmiş ürünler(ürün çeşitleri) : {temizlenmis_urunler}")
```

```python
if orders[3]["is_paid"] == True :
    print("3. sipariş onaylandı.")
else :
    print("3.sipariş için ödeme bekleniyor")
```

```python
#GÜVENLİ SİPARİŞ ÖZETİ(TUPLE)
```

```python
#tuple da immutability özelliği olduğundan değiştirilemez.ekleme ve güncelleme yapılamaz bu da bize daha güvenlikli yapı sağlar.
```

```python
print(f"sipariş özeti : {siparis_ozeti}")
```
